```python
import pandas as pd
import os
```

```python
# Load your CSV
df = pd.read_csv("palm_tree_dataset/train_labels.csv")
```

```python
df
```

```python
# Filter only Palm class (ensure 'class' column exists)
df = df[df['class'].str.lower() == 'palm']  # safer comparison
print(f"Filtered Palm rows: {len(df)}")
```

```python
df
```

```python
df.head(6)
```

```python
# Output label folder
output_dir = "labels/train"
os.makedirs(output_dir, exist_ok=True)
```

```python
# Group by image
for image_name, group in df.groupby('filename'):
    image_width = group.iloc[0]['width']
    image_height = group.iloc[0]['height']
```

```python
yolo_lines = []
for _, row in group.iterrows():
    xmin, ymin, xmax, ymax = row['xmin'], row['ymin'], row['xmax'], row['ymax']
    x_center = (xmin + xmax) / 2.0 / image_width
    y_center = (ymin + ymax) / 2.0 / image_height
    box_width = (xmax - xmin) / image_width
    box_height = (ymax - ymin) / image_height

    yolo_lines.append(f"0 {x_center:.6f} {y_center:.6f} {box_width:.6f} {box_height:.6f}")

# Write to YOLO label file
label_file = os.path.join(output_dir, image_name.replace('.jpg', '.txt'))
with open(label_file, 'w') as f:
    f.write('\n'.join(yolo_lines))


```
